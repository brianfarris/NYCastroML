```python
from scipy import stats
dist = stats.uniform(-1,2)
r = dist.rvs(10)
p = dist.pdf(0.0)
print "r: ",r
print "p: ",p
x = np.linspace(-2, 2, 1000)
plt.plot(x,dist.pdf(x))
plt.show()
```

```python
dist = stats.norm(0,1)
r = dist.rvs(10)
p = dist.pdf(1)
x = np.linspace(-4, 4, 1000)
plt.plot(x,dist.pdf(x))
plt.show()
```

```python
dist = stats.binom(200,0.1)
r = dist.rvs(10)
p = dist.pmf(10)
x = np.arange(-1, 200)
plt.plot(x,dist.pmf(x),linestyle='steps-mid')
plt.xlim(-1,35)
plt.show()
```

```python
from scipy import stats
dist = stats.poisson(5)
x = np.arange(-1,25)
plt.plot(x,dist.pmf(x),linestyle='steps-mid')
plt.xlim(-1,25)
plt.show()
```

```python
dist = stats.cauchy(0,1)
x = np.linspace(-10,10,1000)
plt.plot(x,dist.pdf(x))
plt.xlim(-10,10)
plt.show()
```

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
from matplotlib.patches import Ellipse
from astroML.stats import fit_bivariate_normal
from astroML.stats.random import bivariate_normal

import matplotlib
N = 1000

sigma1 = 2.0
sigma2 = 1.0
mu = [10, 10]
alpha_deg = 45.0
alpha = alpha_deg * np.pi / 180
np.random.seed(0)
X = bivariate_normal(mu, sigma1, sigma2, alpha, N)
fig,axes = plt.subplots(1,2,figsize=(10,5))
f=0.05
X[:int(f * N)] = bivariate_normal((10, 10), 2, 4,45 * np.pi / 180., int(f * N))
x, y = X.T
(mu_nr, sigma1_nr,sigma2_nr, alpha_nr) = fit_bivariate_normal(x, y, robust=False)
(mu_r, sigma1_r,sigma2_r, alpha_r) = fit_bivariate_normal(x, y, robust=True)
axes[0].scatter(x,y,s=2, lw=0, c='k', alpha=0.5)
# Draw elipses showing the fits
for Nsig in [1, 3]:
    # True fit
    E = Ellipse((10, 10), sigma1 * Nsig, sigma2 * Nsig, alpha_deg,ec='k', fc='none')
    axes[0].add_patch(E)

    # Non-robust fit
    E = Ellipse(mu_nr, sigma1_nr * Nsig, sigma2_nr * Nsig,(alpha_nr * 180. / np.pi),ec='k', fc='none', linestyle='dotted')
    axes[0].add_patch(E)

    # Robust fit
    E = Ellipse(mu_r, sigma1_r * Nsig, sigma2_r * Nsig,(alpha_r * 180. / np.pi),ec='k', fc='none', linestyle='dashed')
    axes[0].add_patch(E)
f=0.15
X[:int(f * N)] = bivariate_normal((10, 10), 2, 4,45 * np.pi / 180., int(f * N))
x, y = X.T
(mu_nr, sigma1_nr,sigma2_nr, alpha_nr) = fit_bivariate_normal(x, y, robust=False)
(mu_r, sigma1_r,sigma2_r, alpha_r) = fit_bivariate_normal(x, y, robust=True)
axes[1].scatter(x,y,s=2, lw=0, c='k', alpha=0.5)
# Draw elipses showing the fits
for Nsig in [1, 3]:
    # True fit
    E = Ellipse((10, 10), sigma1 * Nsig, sigma2 * Nsig, alpha_deg,ec='k', fc='none')
    axes[1].add_patch(E)

    # Non-robust fit
    E = Ellipse(mu_nr, sigma1_nr * Nsig, sigma2_nr * Nsig,(alpha_nr * 180. / np.pi),ec='k', fc='none', linestyle='dotted')
    axes[1].add_patch(E)

    # Robust fit
    E = Ellipse(mu_r, sigma1_r * Nsig, sigma2_r * Nsig,(alpha_r * 180. / np.pi),ec='k', fc='none', linestyle='dashed')
    axes[1].add_patch(E)
axes[0].set_xlim(6,14)
axes[1].set_xlim(6,14)
axes[0].set_ylim(6,14)
axes[1].set_ylim(6,14)
plt.show()
```
